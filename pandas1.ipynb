```python
import pandas as pd
df = pd.read_excel('경제활동.xlsx')
df
```

```python
# '시점' 열의 값이 2021인 행을 추출하여 새로운 데이터프레임 생성
df_2021 = df[df['시점'] == 2021]

# 생성된 데이터프레임 확인
df_2021
```

```python
# df_2021 데이터프레임의 행과 열을 바꿉니다 (Transpose).
df_2021_transposed = df_2021.T

# 결과 확인
df_2021_transposed
```

```python
# df_2021 데이터프레임에서 '항목'과 '시점' 열을 제외하고 데이터를 추출합니다.
df_2021_data = df_2021.drop(columns=['항목', '시점'])

# 결과 확인
df_2021_data
```

```python
# df_2021_data 데이터프레임의 행과 열을 바꿉니다.
df_2021_data_transposed = df_2021_data.T

# 결과 확인
df_2021_data_transposed
```

```python
# df_2021에서 '항목' 열의 값을 가져와 새로운 열 이름으로 사용합니다.
new_column_names = df_2021['항목'].values

# df_2021_data_transposed의 열 이름을 변경합니다.
df_2021_data_transposed.columns = new_column_names

# 열 이름이 변경된 데이터프레임을 확인합니다.
df_2021_data_transposed
```

```python
# df_2021_data_transposed에 '연도' 열을 추가하고 값을 2021로 설정합니다.
df_2021_data_transposed['연도'] = 2021

# 결과 확인
df_2021_data_transposed
```

```python
# 처리할 연도 리스트
years_to_process = [2021, 2022, 2023, 2024]

# 각 연도별로 처리된 데이터프레임을 저장할 딕셔너리
yearly_data = {}

for year in years_to_process:
    # 1. 해당 연도의 데이터 추출
    df_year = df[df['시점'] == year]

    # 원본 데이터에 해당 연도 데이터가 없을 경우를 대비
    if df_year.empty:
        print(f"{year}년 데이터가 없습니다.")
        continue

    # 2. '항목'과 '시점' 열을 제외하기 전, 열 이름으로 사용할 '항목' 값을 따로 저장
    df_year_filtered = df_year.reset_index(drop=True)
    column_names = df_year_filtered['항목'].values
    
    # 3. '항목', '시점' 열 제외
    df_year_data = df_year_filtered.drop(columns=['항목', '시점'])

    # 4. 행/열 전환
    df_year_transposed = df_year_data.T

    # 5. 열 이름 설정
    df_year_transposed.columns = column_names

    # 6. '연도' 열 추가
    df_year_transposed['연도'] = year

    # 7. 딕셔너리에 저장
    yearly_data[year] = df_year_transposed
    print(f"✅ {year}년 데이터 처리 완료.")

# 처리된 데이터프레임 중 하나를 확인 (예: 2021년)
if 2021 in yearly_data:
    print("\n--- 2021년 처리 결과 ---")
    display(yearly_data[2021])
# 처리된 데이터프레임 중 하나를 확인 (예: 2022년)
if 2022 in yearly_data:
    print("\n--- 2022년 처리 결과 ---")
    display(yearly_data[2022])
# 처리된 데이터프레임 중 하나를 확인 (예: 2023년)
if 2023 in yearly_data:
    print("\n--- 2023년 처리 결과 ---")
    display(yearly_data[2023])
    
# 처리된 데이터프레임 중 하나를 확인 (예: 2024년)
if 2024 in yearly_data:
    print("\n--- 2024년 처리 결과 ---")
    display(yearly_data[2024])
```

```python
# yearly_data 딕셔너리에 있는 모든 데이터프레임을 하나로 합칩니다.
df_total = pd.concat(yearly_data.values())

# 합쳐진 전체 데이터프레임을 확인합니다.
df_total
```

```python
# df_total 데이터프레임을 '경제활동_통합.csv' 파일로 저장합니다.
# index=True로 설정하여 지역 이름(인덱스)을 파일에 포함시킵니다.
# encoding='utf-8-sig'는 Excel에서 한글이 깨지지 않도록 합니다.
file_name = '경제활동_통합.csv'
df_total.to_csv(file_name, encoding='utf-8-sig')

print(f"✅ 데이터프레임이 '{file_name}' 파일로 성공적으로 저장되었습니다.")
```

```python
# '경제활동생성.csv' 파일을 다시 데이터프레임으로 불러옵니다.
# index_col=0 옵션은 파일의 첫 번째 열을 데이터프레임의 인덱스로 사용하도록 지정합니다.
df_loaded = pd.read_csv('경제활동생성.csv', index_col=0)

# 불러온 데이터프레임의 전체 내용을 확인합니다.
df_loaded
```

```python
# 1. 인덱스 라벨 '계'를 '전국'으로 변경합니다.
df_loaded.rename(index={'계': '전국'}, inplace=True)

# 2. 인덱스의 이름을 '지역'으로 설정합니다.
df_loaded.index.name = '지역'

# 변경된 데이터프레임을 확인합니다.
df_loaded
```

```python
# 현재 인덱스('지역')를 새로운 열로 변환하고, 기본 숫자 인덱스를 사용합니다.
df_loaded.reset_index(inplace=True)

# 변경된 데이터프레임을 확인합니다.
df_loaded
```

```python
# '실업률'과 '취업률'을 계산하여 새로운 열로 추가합니다.
# 실업률(%) = (실업자 / 경제활동인구) * 100
# 취업률(%) = (취업자 / 경제활동인구) * 100
df_loaded['실업률'] = (df_loaded['실업자 (천명)'] / df_loaded['경제활동인구 (천명)']) * 100
df_loaded['취업률'] = (df_loaded['취업자 (천명)'] / df_loaded['경제활동인구 (천명)']) * 100

# 계산 결과를 소수점 둘째 자리까지 반올림하여 가독성을 높입니다.
df_loaded['실업률'] = df_loaded['실업률'].round(2)
df_loaded['취업률'] = df_loaded['취업률'].round(2)

# 결과 확인
df_loaded
```

```python
# '지역' 열의 값이 '전국'인 데이터만 추출하여 새로운 데이터프레임을 생성합니다.
df_nationwide = df_loaded[df_loaded['지역'] == '전국']

# 추출된 '전국' 데이터프레임을 확인합니다.
df_nationwide
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 운영체제에 맞는 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin': # Mac
    plt.rc('font', family='AppleGothic')
else: # Linux/Colab
    # 나눔고딕 폰트가 설치되어 있어야 합니다.
    plt.rc('font', family='NanumGothic')

# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 1. 기본 Figure와 Axis 생성 (실업률을 주축으로 설정)
fig, ax1 = plt.subplots(figsize=(12, 8))

# 2. 실업률을 선 그래프로 그리기 (ax1, 왼쪽 y축 - 주축)
sns.lineplot(data=df_nationwide, x='연도', y='실업률', color='tomato', marker='s', ax=ax1, label='실업률')
ax1.set_ylabel('실업률 (%)', fontsize=12, color='tomato')
ax1.tick_params(axis='y', labelcolor='tomato')
ax1.set_ylim(0, 5) # 실업률 범위에 맞게 y축 조정

# x축 설정을 명확하게 지정합니다.
years = df_nationwide['연도'].unique() # [2021, 2022, 2023, 2024]
ax1.set_xlabel('연도', fontsize=12)
plt.xticks(ticks=years, labels=[f'{y}년' for y in years])

# 선 그래프 위에 값 표시 (실업률)
for x, y in zip(df_nationwide['연도'], df_nationwide['실업률']):
    ax1.text(x=x, y=y+0.1, s=f'{y:.2f}%', ha='center', color='tomato', fontsize=11)

# 3. 취업률을 선 그래프로 그리기 (ax2, 오른쪽 y축 - 보조축)
ax2 = ax1.twinx() # x축을 공유하는 두 번째 y축 생성
sns.lineplot(data=df_nationwide, x='연도', y='취업률', color='royalblue', marker='o', ax=ax2, label='취업률')
ax2.set_ylabel('취업률 (%)', fontsize=12, color='royalblue')
ax2.tick_params(axis='y', labelcolor='royalblue')
ax2.set_ylim(96, 98) # 취업률 범위에 맞게 y축 조정

# 선 그래프 위에 값 표시 (취업률)
for x, y in zip(df_nationwide['연도'], df_nationwide['취업률']):
    ax2.text(x=x, y=y+0.05, s=f'{y:.2f}%', ha='center', color='royalblue', fontsize=11)

# 4. 그래프 제목 및 범례 설정
plt.title('전국 연도별 취업률 및 실업률 추이', fontsize=16, pad=20)
ax1.grid(True, axis='y', linestyle='--', alpha=0.7)
fig.tight_layout() # 레이아웃 조정

# 범례를 합쳐서 표시
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1+h2, l1+l2, loc='upper left')
ax2.get_legend().remove() # 보조축의 범례는 제거

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 운영체제에 맞는 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin': # Mac
    plt.rc('font', family='AppleGothic')
else: # Linux/Colab
    # 나눔고딕 폰트가 설치되어 있어야 합니다.
    plt.rc('font', family='NanumGothic')

# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# '전국' 데이터를 제외한 지역별 데이터만 추출
df_regional = df_loaded[df_loaded['지역'] != '전국']

# 그래프 크기 설정 (가로로 길게)
plt.figure(figsize=(20, 8))

# Seaborn을 사용하여 연도별 지역별 취업률 막대그래프 생성
sns.barplot(data=df_regional, x='지역', y='취업률', hue='연도')

# 그래프 제목 및 축 레이블 설정
plt.title('연도별/지역별 취업률', fontsize=16)
plt.xlabel('지역', fontsize=12)
plt.ylabel('취업률 (%)', fontsize=12)
plt.xticks(rotation=45) # x축 레이블 회전
plt.ylim(90, 100) # y축 범위를 조정하여 차이를 더 잘 보이게 함
plt.legend(title='연도')
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

# 레이아웃 조정 및 그래프 보여주기
plt.tight_layout()
plt.show()
```
